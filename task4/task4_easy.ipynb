В этой задаче нужно визуализировать *центральную предельную теорему*. Перед выполнением этой задачи настоятельно рекомендуется посмотреть <a href="https://mipt-stats.gitlab.io/courses/ad_fivt/LLN.html">ноутбук</a> с лекции про закон больших чисел. **a).** Пусть $\xi_1, ..., \xi_n$ &mdash; независимые случайные величины из распределения $Exp(\lambda)$. Согласно центральной предельной теореме выполнена сходимость $$Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\longrightarrow} \mathcal{N}(0, 1),$$ где $X_n = \sum\limits_{i=1}^n \xi_i$. Вам нужно убедиться в этом, сгенерировав множество наборов случайных величин и посчитав по каждому из наборов величину $Z_n$ в зависимости от размера набора. 

Сгенерируйте 500 наборов случайных величин $\xi_1^j, ..., \xi_{300}^j$ из распределения $Exp(1)$. По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора. Для этого среднего посчитайте величину $$Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}.$$
Обратите внимание, что нужно взять теоретические математическое ожидание и дисперсию, а не вычислять их по выборке.

```python
import scipy.stats as sps
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(cumsums, expected_value, expected_value_squared, n=300):
    variance = n**2 * expected_value_squared - (n * expected_value)**2
    return (cumsums[:, n-1] - expected_value * n) / variance**0.5
```

```python
samples = np.array([sps.expon(scale=1).rvs(size=300) for _ in range(500)])
cumsums = np.cumsum(samples, axis=1)
zs = f(cumsums, expected_value = 1, expected_value_squared = 2, n = 300)
# len(zs)
```

```python
plt.figure(figsize=(10, 4))

sample = zs


plt.hist(
    sample, 
    bins=20,
    alpha=0.5,
    color='orange'
)

plt.xlabel('значения zs')
plt.ylabel('количество значений')

plt.show()
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

```python
ns = np.array(range(300))
zss = np.array([f(cumsums, 1, 2, i) for i in range(1, 301)])
```

```python
plt.figure(figsize=(12, 8), dpi=80)
for i in range(500):
    plt.scatter(ns, zss[:, i], alpha=0.05)
plt.xlabel(r'n', fontsize=15)
plt.ylabel(r'Z', fontsize=15)
plt.show()
```

Да, сходится к 0.

Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{500,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$, которую тоже постройте на том же графике? Не забудьте сделать легенду.

```python
plt.figure(figsize=(10, 4))

sample = f(cumsums, 1, 2, n=300)


plt.hist(
    sample, 
    bins=20,
    alpha=0.5,
    color='orange'
)

plt.xlabel('значения zs')
plt.ylabel('количество значений')

plt.show()
```

Да, распределение вполне похоже стандартное нормальное.

**b).** Выполните те же действия для распределений $U(0, 1)$ и $Pois(1)$.

```python
samples = np.array([sps.uniform(0, 1).rvs(size=300) for _ in range(500)])
uniform_expected_value = 0.5 #  (1-0)/2
uniform_expected_value_squared = 1/3
cumsums = np.cumsum(samples, axis=1)

zss = np.array([f(cumsums, uniform_expected_value, uniform_expected_value_squared, i) for i in range(1, 301)])

plt.figure(figsize=(12, 8), dpi=80)
for i in range(500):
    plt.scatter(ns, zss[:, i], alpha=0.05)
plt.xlabel(r'n', fontsize=15)
plt.ylabel(r'Z', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 4))

sample = f(cumsums, uniform_expected_value, uniform_expected_value_squared, 300)


plt.hist(
    sample, 
    bins=20,
    alpha=0.5,
    color='orange'
)

plt.xlabel('значения zs')
plt.ylabel('количество значений')

plt.show()
```

```python
samples = np.array([sps.poisson(1).rvs(size=300) for _ in range(500)])
poisson_expected_value = 1
poisson_expected_value_squared = 2
cumsums = np.cumsum(samples, axis=1)

zss = np.array([f(cumsums, poisson_expected_value, poisson_expected_value_squared, i) for i in range(1, 301)])

plt.figure(figsize=(12, 8), dpi=80)
for i in range(500):
    plt.scatter(ns, zss[:, i], alpha=0.05)
plt.xlabel(r'n', fontsize=15)
plt.ylabel(r'Z', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 4))

sample = f(cumsums, poisson_expected_value, poisson_expected_value_squared, 300)


plt.hist(
    sample, 
    bins=20,
    alpha=0.5,
    color='orange'
)

plt.xlabel('значения zs')
plt.ylabel('количество значений')

plt.show()
```

Да, распределения вполне похожи на стандартное нормальное.
